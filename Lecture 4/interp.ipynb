```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
xs = np.linspace(0, 10, 10)
ys = np.array([0, 3, 2, 1.5, -2, -3, 4, 4.5, 3.5, 2])

# plt.scatter(xs, ys)
plt.plot(xs, ys, 'o-')
```

```python
xs_interp = np.linspace(0, 10, 100)
from scipy.interpolate import lagrange

# Lagrange polynomial interpolation
poly = lagrange(xs, ys)
ys_interp = poly(xs_interp)


plt.plot(xs, ys, 'o-')
plt.plot(xs_interp, ys_interp)
```

```python
from scipy.interpolate import BSpline
from scipy.interpolate import splrep

# Second order B-spline interpolation
tck = splrep(xs, ys, k=3)  # k=2 for second order (quadratic) B-spline
ys_bspline = BSpline(*tck)(xs_interp)

plt.plot(xs, ys, 'o-', label='Original data')
plt.plot(xs_interp, ys_bspline, label='B-spline interpolation (2nd order)')
plt.legend()
# plt.show()
```
